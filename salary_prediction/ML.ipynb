```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# fetching data
df = pd.read_csv("/content/drive/MyDrive/ML/Dataset/Salary_Data.csv")
df.head()
```

```python
# Splitting data
Y = df["Salary"].to_numpy().reshape(-1, 1)
X = df.drop(columns=["Salary"]).to_numpy()
```

```python
# spitting data into training setand testing set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=24)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# predicting using the given model
y_pred = model.predict(X_test)
```

```python
# Model parameters
print(np.mean(y_test - y_pred))
```
